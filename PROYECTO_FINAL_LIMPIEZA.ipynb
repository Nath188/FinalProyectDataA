# Análisis de texto


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
print(stopwords.words('spanish'))
```

```python
textos_encuestas=''
for row in df_titulos_2019['concepto']:
  textos_encuestas=textos_encuestas +''+row
```

```python
stopwords_list=stopwords.words('spanish')+ ['mile', ' pea', 'dane', 'etc']
```

```python
textos_encuestas= ' '.join([x for x in textos_encuestas.strip().lower().split(' ') if x not in stopwords_list and len(x)>4])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import socket
```

```python
wc = WordCloud(background_color='white')
wc.generate(textos_encuestas)

plt.axis("off")
plt.imshow(wc, interpolation ='bilinear')
plt.show()
```
